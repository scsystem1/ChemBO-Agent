```python
import pandas as pd
from pathlib import Path
from gauche.dataloader import ReactionLoader

loader = ReactionLoader()
loader.load_benchmark("SuzukiMiyaura")

X = loader.features.copy()
y = pd.DataFrame(loader.labels, columns=["yield"])

df = pd.concat(
    [X.reset_index(drop=True), y.reset_index(drop=True)],
    axis=1,
)

print(df.shape)
print(df.head())

out_dir = Path("/home/sjtu/LLMBO/ChemBO-Agent/data/collect_data_from_gauche")
out_dir.mkdir(parents=True, exist_ok=True)

out_path = out_dir / "suzuki_miyaura_5760.csv"
df.to_csv(out_path, index=False)

print(f"saved to: {out_path}")
```
